```python
for j in [100*i+100 for i in range(9)]:
    fig, axes = plt.subplots(1,2)
    im = axes[0].imshow(l.decoded_future[j,:,:,0], vmin=-0.2, vmax=1)
    im = axes[1].imshow(l.data[j,:,:,0], vmin=-0.2, vmax=1)

    fig.subplots_adjust(right=0.8)
    cbar_ax = fig.add_axes([0.85, 0.15, 0.05, 0.7])
    fig.colorbar(im, cax=cbar_ax)

    plt.show()
```
