```python
import sys

sys.path.append("../..")
import numpy as np
import matplotlib.pyplot as plt

from scripts.neuralnetwork.rnn import RNN
```

```python
# Name of the trained RNN to be uploaded
model_name = "VDP_rnn"

# Instantiate the object
rnn = RNN(model_name=model_name)
```

```python
# Load the test dataset
test = np.load("../../dataset/VDP_testset/VDP_testset_0.npz")["data"][0]

# Predict the future using the rnn
forecast = rnn.predict_future(
    starting_sequence=test[: rnn.window_size], length=(250 - rnn.window_size)
)

# Concatenate the predicted future and the past data
forecast_full = np.concatenate((test[: rnn.window_size], forecast), axis=0)
```

```python
# Plot the prediction vs the true values

for i in range(2):
    plt.plot(forecast_full[:, i])
    plt.plot(test[:, i])
    plt.legend(["forecasted", "true"])
    # plt.vlines(r.window_size, ymin=np.min(test[:,i]),ymax=np.max(test[:,i]), colors='red')
    plt.show()
```
