```python
import sys

sys.path.append("../..")
import numpy as np
import matplotlib.pyplot as plt

from scripts.neuralnetwork.rnn import RNN
```

```python
# name of the trained RNN to be uploaded
model_name = "FN_rnn"

# Instanciate the object
rnn = RNN(model_name=model_name)
```

```python
# Load the test dataset
test = np.load("../../dataset/FN_train_encoded.npy")[0]

# Predict the future using the rnn
forecast = rnn.predict_future(
    starting_sequence=test[: rnn.window_size], length=(1000 - rnn.window_size)
)

# Concatenate the predicted data and the starting ones (the first window_size time steps from test)
forecast_full = np.concatenate((test[: rnn.window_size], forecast), axis=0)
```

```python
# plot the prediction vs the true values (in the latent space)

for i in range(10):
    plt.plot(test[:, i])
    plt.plot(forecast_full[:, i])
    plt.legend(["Test data", "Forecasted"])
    plt.show()
```
